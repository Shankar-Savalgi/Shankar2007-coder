```python
# Step -1 ( Import basic libraries )
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step - 2 ( Load the data )
df=pd.read_csv("customer_churn.csv")
df
```

```python
# Step - 3 ( EDA - Exploratory Data Analysis )

df.info()
```

**we dont need to remove outliers in Tree algorithms because these also are robust to outliers**

```python
# Step - 4 ( Data Pre-processing )

# Remove the customer id extra columns
df.drop(columns=["customerID"], inplace=True)
```

```python
# Encoding - Convert Categorical values to Numerical values
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in df.columns:
  if df[i].dtype=="object": # If any column is having data type object , Apply Label Encoding to it
    df[i] = le.fit_transform(df[i])
```

```python
df  # Clean data ready for model building
```

```python
# Step - 5 ( Data Splitting for Model Building )

x= df.drop("Churn", axis=1)  # Independent variables
y= df["Churn"]  # Dependent variables

# Divide my data into training and testing data
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split( x , y , test_size=0.2 , random_state=42 )
```

```python
# Step - 6 ( Model Training and Testing )
from sklearn.ensemble import RandomForestClassifier

model_1 = RandomForestClassifier()

# Train the model
model_1.fit(x_train , y_train)

# Ask the model to predict
y_pred = model_1.predict(x_test)

# Evalaute the performance by comparing predicted values with Actual values
from sklearn.metrics import *

print("Accuracy score of the model is :", accuracy_score(y_pred , y_test)*100)
```

```python
# Step - 6 ( Model Training and Testing )
from sklearn.ensemble import RandomForestClassifier

model_1 = RandomForestClassifier(n_estimators=200 , criterion="entropy" , max_depth=10 , min_samples_leaf=5 , min_samples_split=5)

# Train the model
model_1.fit(x_train , y_train)

# Ask the model to predict
y_pred = model_1.predict(x_test)

# Evalaute the performance by comparing predicted values with Actual values
from sklearn.metrics import *

print("Accuracy score of the model is :", accuracy_score(y_pred , y_test)*100)
```

```python
paramaters_grid={ "n_estimators":[50,100,150,200] ,"criterion":["gini","entropy"] ,"max_depth":[5,7,8,9,10,11,12,13,14,15,20] ,"min_samples_leaf":[1,5,10,15,20] , "min_samples_split":[1,5,10,15,20] }

from sklearn.model_selection import GridSearchCV
model=RandomForestClassifier()

grid_search = GridSearchCV( estimator=model , param_grid=paramaters_grid)
grid_search.fit(x_train , y_train)
print("Best hyperparamater values are:", grid_search.best_params_)
```
